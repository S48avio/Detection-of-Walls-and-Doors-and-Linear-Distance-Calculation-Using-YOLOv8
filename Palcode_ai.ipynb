```python
import cv2
import numpy as np
import pytesseract
import re

doors = []
walls = []

# Assume 'results' and 'model' are already defined
for r in results:
    boxes = r.boxes
    for box in boxes:
        cls = int(box.cls[0])
        label = model.names[cls]
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        if label.lower() == "door":
            doors.append(((x1, y1), (x2, y2)))
        elif label.lower() == "wall":
            walls.append(((x1, y1), (x2, y2)))
```

```python
for i, ((x1, y1), (x2, y2)) in enumerate(doors):
    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
    cx, cy = (x1 + x2) // 2, (y1 + y2) // 2
    cv2.putText(image, f"D{i+1}", (cx, cy), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)

cv2_imshow(image)
cv2.waitKey(1)
```

```python
door_index = int(input(f"Enter door number (1-{len(doors)}): ")) - 1
((dx1, dy1), (dx2, dy2)) = doors[door_index]
door_center = np.array([(dx1 + dx2) // 2, (dy1 + dy2) // 2])
door_corners = [(dx1, dy1), (dx1, dy2), (dx2, dy1), (dx2, dy2)]
```

```python
def min_distance_to_walls(point, walls):
    x, y = point
    min_dist = float('inf')
    for (wx1, wy1), (wx2, wy2) in walls:
        x_clamped = np.clip(x, wx1, wx2)
        y_clamped = np.clip(y, wy1, wy2)
        dist = np.linalg.norm(np.array([x, y]) - np.array([x_clamped, y_clamped]))
        min_dist = min(min_dist, dist)
    return min_dist
```

```python
distances = [(pt, min_distance_to_walls(pt, walls)) for pt in door_corners]
distances.sort(key=lambda x: x[1])
closest_two = [distances[0][0], distances[1][0]]
possible_tip = [pt for pt in door_corners if pt not in closest_two]

for pt in door_corners:
    cv2.circle(image, pt, 6, (255, 255, 255), 1)
for pt in closest_two:
    cv2.circle(image, pt, 6, (255, 0, 0), -1)
for pt in possible_tip:
    cv2.circle(image, pt, 6, (0, 0, 255), -1)
cv2.circle(image, tuple(door_center), 6, (255, 0, 255), -1)
```

```python
def raycast_from_point(start, direction, walls, max_distance=1000):
    direction = direction / np.linalg.norm(direction)
    for i in range(1, max_distance):
        probe = start + direction * i
        x_probe, y_probe = int(probe[0]), int(probe[1])
        for (wx1, wy1), (wx2, wy2) in walls:
            if wx1 <= x_probe <= wx2 and wy1 <= y_probe <= wy2:
                return (x_probe, y_probe)
    return None
```

```python
p1, p2 = np.array(closest_two[0]), np.array(closest_two[1])
wall_dir = p2 - p1
wall_dir = wall_dir / np.linalg.norm(wall_dir)
normal_dir = np.array([-wall_dir[1], wall_dir[0]])

hit_1 = raycast_from_point(door_center, normal_dir, walls)
hit_2 = raycast_from_point(door_center, -normal_dir, walls)

closest_mid = np.mean(closest_two, axis=0)
dist1 = np.linalg.norm(np.array(hit_1) - closest_mid) if hit_1 else 0
dist2 = np.linalg.norm(np.array(hit_2) - closest_mid) if hit_2 else 0
final_hit = hit_1 if dist1 > dist2 else hit_2

if final_hit:
    cv2.line(image, tuple(door_center), final_hit, (0, 0, 0), 2)
    cv2.circle(image, final_hit, 5, (0, 255, 255), -1)
```

```python
ocr_text = pytesseract.image_to_string(image)
pattern = r"1\s*cm\s*=\s*(\d+(\.\d+)?)\s*m"
match = re.search(pattern, ocr_text)

if match:
    scale_value_m = float(match.group(1))
    print(f"Scale from image: 1 cm = {scale_value_m} meters")
else:
    print("❌ No scale found (e.g., '1 cm = 5 m'). Please provide:")
    cm = float(input("Enter number of cm: "))
    m = float(input("Enter equivalent in meters: "))
    scale_value_m = m / cm
```

```python
pixels_per_cm = float(input("Enter how many pixels = 1 cm in the image: "))

if final_hit:
    pixel_distance = np.linalg.norm(np.array(final_hit) - door_center)
    cm_distance = pixel_distance / pixels_per_cm
    real_world_distance = cm_distance * scale_value_m
    print(f"\n📐 Distance from door to wall:")
    print(f"→ {pixel_distance:.2f} pixels")
    print(f"→ {cm_distance:.2f} cm")
    print(f"→ {real_world_distance:.2f} meters")
```
